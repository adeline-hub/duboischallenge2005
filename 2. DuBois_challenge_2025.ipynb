####import data

```python
linkk = 'https://raw.githubusercontent.com/ajstarks/dubois-data-portraits/refs/heads/master/challenge/2025/challenge02/data.csv'
df = pd.read_csv(linkk)
print(df.dtypes)
df.reset_index(drop=True, inplace=True)
df.head(2)
```

#### (1st test) create the barplot with seaborn

```python
# Set Seaborn style
sns.set_style("dark")
sns.set_style(rc = {'axes.facecolor': 'oldlace'})
bar_width = 1

ax = sns.barplot(y=df["Date"], x=df["Land"], orient='h', color="r", width=bar_width)

ax.set_title("ACRES OF LAND OWNED BY NEGROES \n IN GEORGIA.", fontsize=14, fontweight='light', fontname="Sans Serif",
             pad=-20, loc="center", color="black")
ax.set(xlabel=None, ylabel=None)

# Text annotation
ax.text(160000, 1, "338,769")
ax.text(500000, 24, "1,062,223")
```
